Melakukan cek tenure dengan age, avg_friend_age dan songsListened. Karena berdasarkan uji statistik korelasi nilai pada ketiga metriks ini memiliki nilai tertinggi

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# fungsi menghitung R-squared
def calculate_r2(df, x_col, y_col='tenure'):
    df = df[[x_col, y_col]].dropna()
    X = df[[x_col]].values
    y = df[y_col].values

    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)

    return r2_score(y, y_pred)

r2_age = calculate_r2(df, 'age')
r2_avg_friend_age = calculate_r2(df, 'avg_friend_age')
r2_songs_listened = calculate_r2(df, 'songsListened')

print("R Result:")
print(f"age: {r2_age:.4f}")
print(f"avg_friend_age: {r2_avg_friend_age:.4f}")
print(f"songsListened: {r2_songs_listened:.4f}")
```

```python
# visualisasi hubungan tenure dengan beberapa variabel
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.scatterplot(data=df, x='age', y='tenure', ax=axes[0]).set(title= f'Tenure vs Age (R Result:{r2_age:.4f})')
sns.scatterplot(data=df, x='avg_friend_age', y='tenure', ax=axes[1]).set(title=f'Tenure vs Avg Friend Age (R Result:{r2_avg_friend_age:.4f})')
sns.scatterplot(data=df, x='songsListened', y='tenure', ax=axes[2]).set(title=f'Tenure vs Songs Listened (R Result:{r2_songs_listened:.4f})')

plt.tight_layout()
plt.show()
plt.show()
```

Hasil Visualisasi:
1. Usia Pengguna vs. Tenure:
 * Semakin tua usia pengguna, cenderung semakin lama mereka bertahan di MusicMap.
 * Ada beberapa outlier pengguna muda dengan tenure tinggi.

2. Rata-rata Usia Teman vs. Tenure:
  * Korelasi positif terlihat, pengguna yang memiliki teman lebih tua cenderung bertahan lebih lama.

3. Jumlah Lagu Didengar vs. Tenure:
  * Semakin banyak lagu yang didengar, semakin tinggi tenure.

### **Interpretasi Data Berdasarkan Gambar**

Data ini membandingkan perilaku pengguna berdasarkan dua faktor utama:
1. **Adopter (0 = Non-Premium, 1 = Premium)**
2. **Good Country (0 = Negara selain US, UK, Jerman, 1 = US, UK, Jerman)**  

Dari tabel, beberapa insight yang bisa diambil:

#### **1. Pengguna Premium Lebih Aktif Dibandingkan Non-Premium**
   - Pengguna premium **(adopter = 1)** mendengarkan lebih banyak lagu, memiliki pengeluaran lebih tinggi, dan lebih aktif di komunitas (lebih banyak posting, playlist, dan shouts) dibandingkan dengan pengguna non-premium.
   - Contoh:  
     - Pengguna premium dari luar "Good Country" mendengarkan **32.906 lagu** vs pengguna non-premium yang hanya **16.482 lagu**.  
     - Hal serupa terjadi untuk pengguna premium dari "Good Country" (34.675 lagu) vs non-premium (18.486 lagu).

#### **2. Pengguna dari "Good Country" Menghabiskan Lebih Banyak Uang untuk Berlangganan**
   - Pengguna premium dari "Good Country" rata-rata menghabiskan **$294,27$**, lebih tinggi dibandingkan pengguna premium dari luar "Good Country" yang hanya **$244,37$**.
   - Bahkan pengguna non-premium dari "Good Country" memiliki pengeluaran yang sedikit lebih tinggi (**$92,76$ vs $82,42$**) dibandingkan pengguna non-premium dari luar "Good Country".

#### **3. Pengguna dari Luar "Good Country" Lebih Aktif dalam Interaksi Komunitas**
   - Meskipun pengguna dari "Good Country" menghabiskan lebih banyak uang, pengguna dari luar "Good Country" lebih aktif dalam komunitas:
     - Pengguna premium dari luar "Good Country" memiliki **112,74 shouts** dibandingkan dengan pengguna premium dari "Good Country" yang hanya **91,05 shouts**.
     - Pengguna non-premium dari luar "Good Country" juga memiliki **lebih banyak posts (5,55 vs 4,55) dan shouts (33,02 vs 22,21)** dibandingkan dengan non-premium dari "Good Country".
